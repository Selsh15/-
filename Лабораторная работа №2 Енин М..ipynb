```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла
data = pd.read_csv('life-expectancy-new.txt', delimiter=',')
print(data)

# Получение списка уникальных стран
countries = data['Entity'].unique()

# Создание графиков для каждой страны
for country in countries:
    country_data = data[data['Entity'] == country]
    years = country_data['Year']
    life_expectancy = country_data['Period_life']

    plt.figure(figsize=(10, 6))
    plt.scatter(years, life_expectancy, alpha=0.5)
    plt.title(f'Life Expectancy in {country}')
    plt.xlabel('Year')
    plt.ylabel('Life Expectancy')
    plt.grid(True)
    plt.show()
```
