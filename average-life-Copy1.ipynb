```python
import matplotlib.pyplot as plt

# Dados
dados = [
    ("United Kingdom", 1738, 35.26),
    ("United Kingdom", 1743, 34.27),
    ("United Kingdom", 1748, 36.47),
    ("Sweden", 1751, 38.43),
    ("Sweden", 1752, 35.41),
    ("Sweden", 1753, 39.74),
    ("United Kingdom", 1753, 39.77),
    ("Sweden", 1754, 37.5),
    ("Finland", 1755, 37.25),
    ("Sweden", 1755, 36.58),
    ("Sweden", 1756, 36.25),
    ("Sweden", 1757, 33.81),
    ("Sweden", 1758, 32.92),
    ("United Kingdom", 1758, 38.12),
    ("Sweden", 1759, 38.21),
    ("Sweden", 1760, 39.71),
    ("Sweden", 1761, 37.83),
    ("Sweden", 1762, 32.49),
    ("Sweden", 1763, 30.46),
    ("United Kingdom", 1763, 35.37)
]

# Organizando os dados por país
paises = {}
for pais, ano, valor in dados:
    if pais not in paises:
        paises[pais] = {"anos": [], "valores": []}
    paises[pais]["anos"].append(ano)
    paises[pais]["valores"].append(valor)

# Criando o gráfico
plt.figure(figsize=(12, 6))

# Plotando as linhas temporais
for pais, dados_pais in paises.items():
    plt.plot(dados_pais["anos"], dados_pais["valores"], marker='o', linestyle='-', label=pais)

# Adicionando barras para destacar os valores mais recentes
anos_ultimos = [dados_pais["anos"][-1] for dados_pais in paises.values()]
valores_ultimos = [dados_pais["valores"][-1] for dados_pais in paises.values()]
plt.bar(anos_ultimos, valores_ultimos, color='gray', alpha=0.5, label='Valores mais recentes')

# Adicionando título e legenda
plt.title('Valores ao longo dos anos por país')
plt.xlabel('Ano')
plt.ylabel('Valor')
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```
